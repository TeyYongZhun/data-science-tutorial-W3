```python
random_array_big = np.random.randint(low = 0, high = 100, size = (6, 3))
print("Original array is: \n", random_array_big)

arr_tranposed = random_array_big.T  # .T means tranpose, commonly used in 2D array
print("\nTranposed array is: \n", arr_tranposed)
```

Question 5

```python
print("Create two random numpy array of size n x p and p x m and then compute the dot product")
n = int(input("Enter the n: "))
p = int(input("Enter the p: "))
m = int(input("Enter the m: "))
print(" ")

array1 = np.random.randint(low = 0 , high = 100, size = (n, p))
print("array1 is: \n", array1)
print("\n")

array2 = np.random.randint(low = 0 , high = 100, size = (p, m))
print("array2 is: \n", array2)
print("\n")

dot_product = np.dot(array1, array2)
print("The dot product is: \n", dot_product)
```
